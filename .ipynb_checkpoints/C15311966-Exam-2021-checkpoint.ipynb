```python
# Prepare values for training data
labels = train_data.pop('Class').values
data = train_data.values
```

```python
print(train_data.describe())
```

```python
X_train, X_test, y_train, y_test = train_test_split(data,labels, test_size=0.33, random_state=0)

# As mentioned above values should be in similar upper and lower boundires 
# so they can be easily compared to one another
# this is where scaling comes in
# you dont want to interfere with the test data
# so you only use transform on test data so the previously fitted values
# dont get recalculated and reapplied to both train and test data
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```
